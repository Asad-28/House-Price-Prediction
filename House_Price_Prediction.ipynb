## Model Implementation

Creating the Linear Regression Model on the datasets.

```python
X = pd.DataFrame(np.c_[boston['NOX'], boston['RM'], boston['DIS'], boston['PTRATIO'], boston['INDUS']], columns = ['NOX','RM','DIS','PTRATIO','LSTAT' ])
Y = boston['House_price']
print(X)
print(Y)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.33,random_state = 5 )
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
regressor = LinearRegression()
regressor.fit(x_train, y_train)
y_pred = regressor.predict(x_test)
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df.head()
```

## Model Testing

Testing the Model and Plotting the result

```python
print('MAE', metrics.mean_absolute_error(y_test, y_pred))
print('MSE', metrics.mean_squared_error(y_test, y_pred))
print('SRMSE', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('R-square', metrics.r2_score(y_test, y_pred))
```

```python
x1 = np.array(x_test['NOX'])
y1 = np.array(y_pred)
m,b = np.polyfit(x1, y1, 1)
```

```python
plt.plot(x_test['NOX'], m*x_test['NOX'] + b)
plt.plot(x_test['NOX'],y_pred,'o')
plt.xlabel("Nitric Oxide Content")
plt.ylabel("Median Value")
```
